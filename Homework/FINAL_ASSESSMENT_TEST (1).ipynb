```python
# 10. Find the unique elements in an array [5,5,11,11,2,3,4,8,14,14,15,5]

def unique(arr):
    for i in range(len(arr)):
 
        d = 0
        for j in range(0, i):
            if (arr[i] == arr[j]):
                d = 1
                break

        if (d == 0):
            print(arr[i])

arr = [5,5,11,11,2,3,4,8,14,14,15,5]
unique(arr)
```

### Q1: What is Machine Leanring?

### Q2: What are the key principles of 'Supervised' Machine learning?

The training dataset is labeled,  By analyzing a known training dataset,The system can provide targets for any new input after sufficient training. It can also compare its output with the correct, intended output to find errors and modify the model accordingly.
### Q3: With regards to Machine Learning Model Fitting, what do we mean my Underfitting and Overfitting?
**Overfitting**: Good performance on the training data, poor generliazation to other data.
 **Why** : The model is too good and complicated for the problem, it learns the detail and noise in the training data to the extent that it negatively impacts the performance of the model on new data.

**Solutions:**

Use a resampling technique to estimate model accuracy.
Hold back a validation dataset.
Use a simplier model
Use Regularization to penalize the weight to avoid overfitting
Adding more data

**Underfitting**: Poor performance on the training data and poor generalization to other data

**Why**: The model is too simple, can't learn the pattern form the data.it is unable to capture the relationship between the input and output variables accurately

**Solutions:**

Increasing the model complexity
Reducing regularization
Adding features to training data


### Q4: Explain how do we ususally split data in preparation for the Machine Learning process?
Usually you split your data into three dataset, train dataset, test dataset, validation dataset, use train dataset to fit the model, use validation dataset to do evaluation to see model performance, meanwhile you can try to do tuning model hyperparameters, finally use test dataset on your final model , you can see accuracy of the model after completing the training phase
